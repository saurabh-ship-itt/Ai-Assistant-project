```python
# converting the date from DD/MM/YY to YYYY-MM-DD
data_info=pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(data_info)
```

```python
time_info=air_quality_data['Time']
print(time_info)
```

```python
time_info=time_info.apply(lambda x: x.replace('.',':'))
print(time_info)
```

```python
# combing 2 series to a pandas dataframe
date_time=pd.concat([data_info, time_info], axis=1)
date_time.columns = ['Date', 'Time']
print(date_time)
```

```python
# Combining Date and Time columns into a single datetime column 'ds'
date_time['ds'] = pd.to_datetime(date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str))
print(date_time)
```

```python
data=pd.DataFrame()
```

```python
 data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
#training the prophet model
model=Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```
